```python
import pandas as pd
from matplotlib import pyplot as plt
import seaborn as sns
from plotly import express as px
```

```python
batter = pd.read_csv("../data/processed/batter_stats.csv")
bowler = pd.read_csv("../data/processed/bowler_stats.csv")
```

```python
top_knocks = (
    batter.sort_values(["runs", "balls"], ascending=[False, True])
    [["batter", "runs", "balls"]]
    .head(10)
    .reset_index(drop=True)
)

top_knocks.index = top_knocks.index + 1

print(top_knocks)
```

```python
hs = (
    batter.sort_values(["runs", "balls"], ascending=[False, True])
    [["batter", "runs", "balls"]]
    .head(10)
    .reset_index(drop=True)
)

# create unique x values internally
hs["x_id"] = [f"{name}_{i}" for i, name in enumerate(hs["batter"])]

fig = px.bar(
    hs,
    x="x_id",
    y="runs",
    title="Highest Individual Scores in IPL History",
    text="runs",
    hover_data=["batter", "balls"]
)

# show only batter names on axis
fig.update_layout(
    xaxis_title="Batter",
    yaxis_title="Runs",
    xaxis=dict(
        tickmode="array",
        tickvals=hs["x_id"],
        ticktext=hs["batter"]
    )
)
fig.update_xaxes(tickangle=315)
fig.show()
```

```python
best_figures = (
    bowler.assign(figures=bowler["wkts"].astype(str) + "/" + bowler["runs"].astype(str))
    .sort_values(by=["wkts", "runs"], ascending=[False, True])
    [["bowler", "figures", 'balls']]
    .head(10)
    .reset_index(drop=True)
)

print(best_figures)
```

```python
top_batters = batter["batter"].value_counts().head(10).reset_index()
top_batters.columns = ["batter", "matches"]

fig = px.bar(
    top_batters,
    x="batter",
    y="matches",
    title="Most Batting Innings",
    text="matches"
)
fig.update_xaxes(tickangle=315)
fig.show()
```

```python
top_bowlers = bowler["bowler"].value_counts().head(10).reset_index()
top_bowlers.columns = ["bowler", "matches"]

fig = px.bar(
    top_bowlers,
    x="bowler",
    y="matches",
    title="Most Bowling Innings",
    text="matches"
)
fig.update_xaxes(tickangle=315)

fig.show()
```

```python
most_runs = (
    batter.groupby("batter")["runs"]
    .sum()
    .sort_values(ascending=False)
    .head(10)
    .reset_index()
)

fig = px.bar(
    top_batters,
    x="batter",
    y="runs",
    title="Most Runs in IPL history",
    text="runs"
)
fig.update_xaxes(tickangle=315)
fig.show()
```

```python
most_wkts = (
    bowler.groupby("bowler")["wkts"]
    .sum()
    .sort_values(ascending=False)
    .head(10)
    .reset_index()
)

fig = px.bar(
    most_wkts,
    x="bowler",
    y="wkts",
    title="Most Wickets in IPL history",
    text="wkts"
)
fig.update_xaxes(tickangle=315)
fig.show()
```

```python
best_bowling = (
    bowler.sort_values(by=['wkts', 'runs'], ascending=[True, False]).tail(10).reset_index()
)
best_bowling["figures"] = (best_bowling["wkts"].astype(str) + "/" + best_bowling["runs"].astype(str))

fig = px.bar(
    best_bowling,
    y="bowler",
    x="runs",
    text="figures",
    orientation="h",
    title="Top IPL Wicket-Takers"
)

fig.show()
```
