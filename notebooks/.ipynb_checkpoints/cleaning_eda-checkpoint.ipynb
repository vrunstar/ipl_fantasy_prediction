```python
import pandas as pd
from matplotlib import pyplot as plt
import seaborn as sns
from plotly import express as px
```

```python
batter = pd.read_csv("../data/processed/batter_stats.csv")
bowler = pd.read_csv("../data/processed/bowler_stats.csv")
```

```python
batter.groupby("batter")['runs'].sum().sort_values(ascending=False).head(10)
```

```python
bowler.groupby("bowler")['wkts'].sum().sort_values(ascending=False).head(10)
```

```python
top_batters = batter["batter"].value_counts().head(10).reset_index()
top_batters.columns = ["batter", "matches"]

fig = px.bar(
    top_batters,
    y="batter",
    x="matches",
    title="Top 10 Batters by Number of Match Entries",
    text="matches"
)

fig.update_layout(xaxis_title="Batter", yaxis_title="Match Entries")
fig.show()
```
